# Web scraping a table 

```python
# importing required libraries
```

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# web scraping data from the below URL
url="https://www.nfl.com/standings/league/2022/REG"
```

```python
# extracting the html from the above link
res=requests.get(url)
```

```python
res
```

```python
# from the responce 200 it is clear that extraction of HTML is sucessful.
```

```python
# extracting the html in the same same format it has in text
soup = BeautifulSoup(res.text,"lxml")
```

```python
table=soup.find("table",{"summary": "Standings - Detailed View"})
```

```python
headers=[]
```

```python
for i in table.find_all("th"):
    headers.append(i.text.strip())
```

```python
headers
```

```python
df=pd.DataFrame(columns=headers)
```

```python
df.columns
```

```python
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```
